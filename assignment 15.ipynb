## 1.GET and POST

```python
GET: To request data from the server
POST: The post() method sends a post request to the specified url.
```

## 2.request used in flask

```python
The flask application handles a request,it creates a Request object based on the environmnet it received from the WSGI server.
Because a worker handles only one request at a time, the request data can be considered global to that worker during that 
Request.
```

## 3.redirect()

```python
Flask redirect is dfined as a function or utility in flask which allows developers to redirect users to a specified URL and 
assign a specified status code. When this function is called, a response object is returned, and the redirection happens to 
the target location with the status code.
```

## 4.template in flask

```python
Templates are filesthat contain static data as well as placeholders for dynamic data. A template is rendered with specific data
to produce a final document. Flask uses the Jinja template library to render templates.In your applications,you will use 
template to render HTML which will display in the users browser.
```

## render_template

```python
render_template is used to generate output from a template file based on the Jinja2 engine that is found in the applications 
templates folder.Note that render_template is typically imported directly from the flask package instead of from flask.
```

## 5.API postman 

```python
import requests
import json

url = "https://selectapi.datascope.refinitiv.com/RestApi/v1/Authentication/RequestToken"

payload = json.dumps({
    Credentials: {
    "Username": "DSS_USERNAME",
    "Password": "DSS_PASSWORD"
  }
})
headers = {
  'Prefer': 'respond-async',
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```
